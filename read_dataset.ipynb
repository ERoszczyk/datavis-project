```python
departure_delay_origin = merged_flights_subset_df.groupby('ORIGIN_AIRPORT')['DEPARTURE_DELAY'].mean().reset_index()
arrival_delay_destination = merged_flights_subset_df.groupby('DESTINATION_AIRPORT')['ARRIVAL_DELAY'].mean().reset_index()

airports_df = pd.merge(airports_df, departure_delay_origin, left_on='IATA_CODE', right_on='ORIGIN_AIRPORT', how='left')
airports_df = pd.merge(airports_df, arrival_delay_destination, left_on='IATA_CODE', right_on='DESTINATION_AIRPORT', how='left')
```

```python
airports_df['DEPARTURE_DELAY'] = airports_df['DEPARTURE_DELAY']
airports_df = airports_df.drop(columns=['ORIGIN_AIRPORT', 'DESTINATION_AIRPORT'])
airports_df.fillna(0, inplace=True)
airports_df
```

```python
airports_df['AVERAGE_DELAY'] = airports_df[['ARRIVAL_DELAY', 'DEPARTURE_DELAY']].mean(axis=1)

airports_df
```

```python
bins = [-np.inf, -15, -5, 5, 15, np.inf]
labels = ['Very Early', 'Early', 'On Time', 'Late', 'Very Late']
airports_df['DELAY_CATEGORY'] = pd.cut(airports_df['AVERAGE_DELAY'], bins=bins, labels=labels)
airports_df
```
