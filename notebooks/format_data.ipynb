### Changing time to hh:mm:ss format

```python
from datetime import time

def split_time(time_hhmm):
    """
    Splits a time value in the format hhmm.0 into hours and minutes.

    Args:
        time (float): Time value represented as a float (e.g., 1926.0).

    Returns:
        datetime.time: An object representing the time extracted from the time value.
    """
    hours = int(time_hhmm) // 100
    minutes = int(time_hhmm) % 100
    return time(hours, minutes)
```

```python
# Because of this Error: "ValueError: hour must be in 0..23", I checked if there is 'ArrTime' value > 2359

result = flights_delay_dataset[flights_delay_dataset['ArrTime'] > 2359]

# It turned out there is a value 2400, which in datetime.time format would be 00:00, so I replaced it

flights_delay_dataset['ArrTime'] = flights_delay_dataset['ArrTime'].replace(2400, 0)
features_added_dataset['ArrTime'] = features_added_dataset['ArrTime'].replace(2400, 0)
```

```python
# For flights_delay_dataset

# flights_delay_dataset['DepHour'] = flights_delay_dataset['DepTime'].apply(lambda x: pd.Series(split_time(x)))
flights_delay_dataset['DepHour'] = flights_delay_dataset['DepTime'].apply(split_time)
flights_delay_dataset['ArrHour'] = flights_delay_dataset['ArrTime'].apply(split_time)
flights_delay_dataset['CRSDepHour'] = flights_delay_dataset['CRSDepTime'].apply(split_time)
flights_delay_dataset['CRSArrHour'] = flights_delay_dataset['CRSArrTime'].apply(split_time)
flights_delay_dataset['WheelsOffHour'] = flights_delay_dataset['WheelsOff'].apply(split_time)
flights_delay_dataset['WheelsOnHour'] = flights_delay_dataset['WheelsOn'].apply(split_time)
```

```python
# For features_added_dataset

features_added_dataset['DepHour'] = features_added_dataset['DepTime'].apply(split_time)
features_added_dataset['ArrHour'] = features_added_dataset['ArrTime'].apply(split_time)
```
